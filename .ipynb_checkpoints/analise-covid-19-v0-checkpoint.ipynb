# 3. Por Estado

```python
estado['state'].unique()
```

```python
estado.dropna(axis=0, inplace=True)
estado.loc[estado['state'] == 'pernambuco', 'state'] = 'pe'
estado.loc[estado['state'].isin(['sÃ£o pauo', 'sÃ£o paulo', 'guaruja', 'sao paulo']), 'state'] = 'sp'
estado.loc[estado['state'] == 'alagoas', 'state'] = 'al'
estado.loc[estado['state'] == 'santa catarina', 'state'] = 'sc'
estado.loc[estado['state'] == 'bahia', 'state'] = 'ba'
estado.loc[estado['state'] == 'maranhao', 'state'] = 'ma'
estado.loc[estado['state'].isin(['minas gerais', 'belo horizonte']), 'state'] = 'mg'
estado.loc[estado['state'] == 'paranÃ¡', 'state'] = 'pr'
estado.loc[estado['state'] == 'rio de janeiro', 'state'] = 'rj'
estado.loc[estado['state'].isin(['vitoria', 'espÃ­rito santo']), 'state'] = 'es'
estado.loc[estado['state'] == 'cearÃ¡', 'state'] = 'ce'
estado.loc[estado['state'] == 'goias', 'state'] = 'go'
```

```python
estado_2019 = estado[(estado['year'] == 2019) & (estado['week'] <= 10)].sort_values('state')
estado_2020 = estado[estado['year'] == 2020].sort_values('state')
```

```python
estado_2019[estado_2019['state'] == 'ac']
```

```python
estado_2019['state'].unique()
```

```python
estado_2020['state'].unique()
```

```python
from itertools import product
n = list(product(range(7), range(4)))[:-1]

state_ls = list(map(lambda x: x.upper(), estado_2020['state'].unique()))

fig, ax = plt.subplots(7, 4, figsize=(20, 20))
for i, state in enumerate(state_ls):
    sns.lineplot(x='week',
                 y='media_tpv',
                 data=estado_2020[estado_2020['state'] == state],
                 ax=ax[n[i][0], n[i][1]],
                 label=2020)
    
    if state in ['ap', 'rr']:
        pass
    else:
        sns.lineplot(x='week',
                     y='media_tpv',
                     data=estado_2019[estado_2019['state'] == state],
                     ax=ax[n[i][0], n[i][1]],
                     label=2019)
    
    ax[n[i][0], n[i][1]].set_title(state)
    
plt.subplots_adjust(hspace=0.6, wspace=0.4)
# fig.delaxes(ax[2, 1]);
```
