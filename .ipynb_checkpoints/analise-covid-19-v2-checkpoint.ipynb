```python
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
pd.options.display.max_rows = 300
```

```python
df = pd.read_csv('data/cielo.csv', parse_dates=['dia']).sort_values('id').dropna(axis=0).reset_index(drop=True)
df['mes'] = df['dia'].apply(lambda x: x.month)
df['dia_da_semana'] = df['dia'].apply(lambda x: x.isoweekday())
df['semana'] = df['dia'].apply(lambda x: x.isocalendar()[1])
```

```python
capitais = ['centro',
            'rio branco',
            'maceió',
            'maceio'
            'macapá',
            'macapa',
            'manaus',
            'salvador',
            'fortaleza',
            'brasília',
            'brasilia',
            'vitória',
            'vitoria',
            'goiânia',
            'goiania',
            'são luís',
            'sao luis',
            'são luis',
            'são luiz',
            'cuiabá',
            'cuiaba',
            'campo grande',
            'belo horizonte',
            '  belo horizonte',
            'belém',
            'belem',
            'joão pessoa',
            'joao pessoa',
            'curitiba',
            'curitiba ',
            'recife',
            'teresina',
            'rio de janeiro',
            'rio  de janeiro',
            'rio de jeneiro',
            'rj',
            'natal',
            'porto alegre',
            'porto velho',
            'boa vista',
            'florianópolis',
            'florianopolis',
            'são paulo',
            'sao paulo',
            'sao paulo ',
            'são paulo ',
            'aracaju',
            'palmas']

df['capital'] = df['cidade'].isin(capitais)
```

```python
map_dept = {6: 'VESTUÁRIO',
            8: 'CALÇADOS / BOLSAS / MALAS',
            9: 'CAMA / MESA / BANHO / UTILIDADES DOMÉSTICAS',
            25: 'ARTIGOS',
            36: 'ÓTICA',
            45: 'MÓVEIS'}

df['dept_cat'] = df['department_id'].map(map_dept)
```

```python
# df
```

# 1. Brasil

```python
df_brasil = pd.DataFrame(df.groupby('dia')['tot_venda'].mean()).reset_index()

df_brasil['mes'] = df_brasil['dia'].apply(lambda x: x.month)
df_brasil['dia_da_semana'] = df_brasil['dia'].apply(lambda x: x.isoweekday())
df_brasil['semana'] = df_brasil['dia'].apply(lambda x: x.isocalendar()[1])

dia_da_semana_dict = {1: 'S', 2: 'T', 3: 'Q', 4: 'Q', 5: 'S', 6: 'S', 7: 'D'}
df_brasil['dia_da_semana'] = df_brasil['dia_da_semana'].map(dia_da_semana_dict)

df_brasil.loc[df_brasil['mes'] == 2, 'semana'] -= 4
df_brasil.loc[df_brasil['mes'] == 3, 'semana'] -= 8

df_brasil['semana_mod'] = np.tile(np.append(df_brasil.loc[df_brasil['mes'] == 1, 'semana'].values[1:], 4), 3)

df_diff = df_brasil.groupby(['mes', 'semana_mod'])['tot_venda'].sum().reset_index()
```

```python
# df_brasil
```

```python
# df_diff
```

```python
diff = {}
diff['mar_jan_1'] = \
str(round(100 * (df_diff.loc[6, 'tot_venda'] / df_diff.loc[0, 'tot_venda'] - 1), 2)).replace('.', ',') + '%'
diff['mar_jan_2'] = \
str(round(100 * (df_diff.loc[7, 'tot_venda'] / df_diff.loc[1, 'tot_venda'] - 1), 2)).replace('.', ',') + '%'
diff['mar_jan_3'] = \
str(round(100 * (df_diff.loc[8, 'tot_venda'] / df_diff.loc[2, 'tot_venda'] - 1), 2)).replace('.', ',') + '%'

diff['mar_fev_1'] = \
str(round(100 * (df_diff.loc[6, 'tot_venda'] / df_diff.loc[3, 'tot_venda'] - 1), 2)).replace('.', ',') + '%'
diff['mar_fev_2'] = \
str(round(100 * (df_diff.loc[7, 'tot_venda'] / df_diff.loc[4, 'tot_venda'] - 1), 2)).replace('.', ',') + '%'
diff['mar_fev_3'] = \
str(round(100 * (df_diff.loc[8, 'tot_venda'] / df_diff.loc[5, 'tot_venda'] - 1), 2)).replace('.', ',') + '%'
```

```python
# diff
```

```python
df_brasil_jan = df_brasil[df_brasil['mes'] == 1].reset_index(drop=True)
df_brasil_fev = df_brasil[df_brasil['mes'] == 2].reset_index(drop=True)
df_brasil_mar = df_brasil[df_brasil['mes'] == 3].reset_index(drop=True)

fig, ax = plt.subplots(figsize=(12, 4))

sns.lineplot(x=range(17), y='tot_venda', data=df_brasil_jan, color='0.6', lw=3, ax=ax)
sns.lineplot(x=range(17), y='tot_venda', data=df_brasil_fev, color='0.4', lw=3, ax=ax)
sns.lineplot(x=range(17), y='tot_venda', data=df_brasil_mar, lw=3, ax=ax)

y_max = ax.get_ylim()[1]

ax.set_xlim(0, 17)
ax.set_xticks(range(17))
ax.set_xticklabels(df_brasil_jan['dia_da_semana'])
ax.axvline(6.5, ymin=0, ymax=200, color='0', ls=':')
ax.axvline(13.5, ymin=0, ymax=200, color='0', ls=':')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.text(1, y_max, '1ª semana', size=15)
ax.text(1, y_max-250, '\u0394(mar-jan) = {}'.format(diff['mar_jan_1']), size=15, color='red')
ax.text(1, y_max-500, '\u0394(mar-fev) = {}'.format(diff['mar_fev_1']), size=15, color='green')
ax.text(8, y_max, '2ª semana', size=15)
ax.text(8, y_max-250, '\u0394(mar-jan) = {}'.format(diff['mar_jan_2']), size=15, color='red')
ax.text(8, y_max-500, '\u0394(mar-fev) = {}'.format(diff['mar_fev_2']), size=15, color='green')
ax.text(15, y_max, '3ª semana', size=15)
ax.text(15, y_max-250, '\u0394(mar-jan) = {}'.format(diff['mar_jan_3']), size=15, color='red')
ax.text(15, y_max-500, '\u0394(mar-fev) = {}'.format(diff['mar_fev_3']), size=15, color='red')
ax.set_ylabel('VENDAS MÉDIAS', size=12)
ax.annotate('Jan/20', (16.1, df_brasil_jan.loc[16, 'tot_venda']), size=12)
ax.annotate('Fev/20', (16.1, df_brasil_fev.loc[16, 'tot_venda']), size=12)
ax.annotate('Mar/20', (16.1, df_brasil_mar.loc[16, 'tot_venda']), size=12)

plt.savefig('imagens/brasil.png');
```

```python
estados = df['state'].unique()
```

```python
df_estados = pd.DataFrame(df.groupby(['state', 'dia'])['tot_venda'].mean()).reset_index()

df_estados['mes'] = df_estados['dia'].apply(lambda x: x.month)
df_estados['dia_da_semana'] = df_estados['dia'].apply(lambda x: x.isoweekday())
df_estados['semana'] = df_estados['dia'].apply(lambda x: x.isocalendar()[1])
```

```python
df_estados_dict = {}
df_diff = {}
for e in estados:
    if len(df_estados[df_estados['state'] == e]) < 51:
        continue
    else:
        df_estados_dict[e] = df_estados[df_estados['state'] == e].reset_index(drop=True)

        dia_da_semana_dict = {1: 'S', 2: 'T', 3: 'Q', 4: 'Q', 5: 'S', 6: 'S', 7: 'D'}
        df_estados_dict[e]['dia_da_semana'] = df_estados_dict[e]['dia_da_semana'].map(dia_da_semana_dict)

        df_estados_dict[e].loc[df_estados_dict[e]['mes'] == 2, 'semana'] -= 4
        df_estados_dict[e].loc[df_estados_dict[e]['mes'] == 3, 'semana'] -= 8

        df_estados_dict[e]['semana_mod'] = \
        np.tile(np.append(df_estados_dict[e].loc[df_estados_dict[e]['mes'] == 1, 'semana'].values[1:], 4), 3)
        
        df_diff[e] = df_estados_dict[e].groupby(['mes', 'semana_mod'])['tot_venda'].sum().reset_index()
```
